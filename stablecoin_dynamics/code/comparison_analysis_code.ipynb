## 4.1 Cross-Exchange Basis Gap

### Definition

For each stablecoin $ s \in \{\text{USDT}, \text{USDC}\} $, define the exchange gap:

$$
\Delta_s(t)
=
\text{Basis}_{s,b}(t)
-
\text{Basis}_{s,k}(t)
$$

where:

- $\text{Basis}_{s,b}(t)$ = stablecoin basis on **Binance**
- $\text{Basis}_{s,k}(t)$ = stablecoin basis on **Kraken**

In code:

$$
\text{usdt\_gap} = \text{usdt}_b - \text{usdt}_k
$$

$$
\text{usdc\_gap} = \text{usdc}_b - \text{usdc}_k
$$

---

### Economic Meaning

$$
\Delta_s(t) > 0
\quad \Rightarrow \quad
\text{Binance basis is larger}
$$

$$
\Delta_s(t) < 0
\quad \Rightarrow \quad
\text{Kraken basis is larger}
$$

This measures **exchange-level fragmentation**.

If markets were perfectly integrated:

$$
\Delta_s(t) \approx 0
$$

Persistent non-zero values indicate:

- Liquidity segmentation  
- Different clienteles  
- Fiat on/off-ramp frictions  
- Arbitrage capital constraints  

---

### Statistical Summary

We compute:

$$
\text{mean}(\Delta_s)
$$

$$
\text{std}(\Delta_s)
$$

$$
p95\big(|\Delta_s|\big)
$$

$$
p99\big(|\Delta_s|\big)
$$

$$
\max\big(|\Delta_s|\big)
$$

These statistics measure:

- **Mean (bps)**  structural exchange bias  
- **Standard deviation (bps)**  volatility of fragmentation  
- **p95 / p99 absolute (bps)**  stress-level divergence  
- **Maximum absolute (bps)**  extreme dislocation  

---

### Interpretation

If:

$$
\text{mean}(\Delta_s) \neq 0
$$

then one exchange systematically prices the stablecoin richer than the other.

Large values of:

$$
p99\big(|\Delta_s|\big)
$$

indicate that during stress events, exchange pricing diverges materially.

Comparing USDT vs USDC gaps reveals which stablecoin exhibits greater cross-exchange fragmentation and liquidity sensitivity.


```python
# 4.1
gap = pd.DataFrame(index=df.index)

gap["usdt_gap"] = df["usdt_b"] - df["usdt_k"]
gap["usdc_gap"] = df["usdc_b"] - df["usdc_k"]

gap.head()
```

```python
# summary
def summarize_gap(x):
    ax = x.abs()
    return pd.Series({
        "mean_bps": x.mean()*1e4,
        "std_bps": x.std()*1e4,
        "p95_abs_bps": ax.quantile(0.95)*1e4,
        "p99_abs_bps": ax.quantile(0.99)*1e4,
        "max_abs_bps": ax.max()*1e4
    })

gap_stats = pd.concat([
    summarize_gap(gap["usdt_gap"]).rename("USDT"),
    summarize_gap(gap["usdc_gap"]).rename("USDC")
], axis=1)

gap_stats
```

## 4.2 LeadLag (Who Prices Risk First?)

### Objective

We test whether one exchange systematically **leads** the other in pricing stablecoin risk.

If pricing adjusts first on one venue, it suggests informational dominance.

---

### Definition

For lag $\ell \in [-L, L]$, compute the cross-correlation:

$$
\rho(\ell)
=
\text{corr}
\left(
\text{Basis}_{s,b}(t),
\text{Basis}_{s,k}(t + \ell)
\right)
$$

where:

- $\text{Basis}_{s,b}(t)$ = Binance basis
- $\text{Basis}_{s,k}(t)$ = Kraken basis
- $\ell$ = time lag (in minutes)

We evaluate:

$$
\ell \in [-60, 60]
$$

---

### Interpretation of Lags

- If the maximum correlation occurs at:

$$
\ell > 0
$$

then **Binance leads** (Kraken adjusts later).

- If the maximum occurs at:

$$
\ell < 0
$$

then **Kraken leads**.

- If the maximum occurs at:

$$
\ell = 0
$$

then pricing adjusts **simultaneously**.

---

### Economic Meaning

Leadlag structure reflects:

- Liquidity depth  
- Arbitrage speed  
- Institutional dominance  
- Fiat on/off-ramp influence  

If Binance leads:

 Global liquidity hub dominance  

If Kraken leads:

 Strong fiat client flow influence  

If symmetric:

 Highly integrated markets  

---

### Interpretation of Your Plot

The peak correlation occurs near:

$$
\ell = 0
$$

This suggests:

- Strong integration between exchanges  
- Rapid arbitrage  
- Minimal persistent information delay  

The high correlation level ( 0.94+) indicates:

$$
\text{Basis}_{b}(t)
\approx
\text{Basis}_{k}(t)
$$

up to very small timing differences.

This implies that cross-exchange fragmentation is mostly **level-based**, not timing-based.


```python
# 4.2
def lead_lag_corr(x, y, max_lag=60):
    lags = range(-max_lag, max_lag+1)
    out = []

    for l in lags:
        if l >= 0:
            c = x.corr(y.shift(-l))
        else:
            c = x.shift(l).corr(y)

        out.append(c)

    return pd.Series(out, index=lags)

# usdt
ll_usdt = lead_lag_corr(df["usdt_b"], df["usdt_k"], max_lag=60)
ll_usdt.plot(title="Lead-lag USDT (Binance vs Kraken)")
```

```python
# usdc
ll_usdc = lead_lag_corr(df["usdc_b"], df["usdc_k"], max_lag=60)
ll_usdc.plot(title="Lead-lag USDC")
```

```python
# 5.1 Stability Ranking
def stability_metrics(x: pd.Series):
    ax = x.abs()
    return pd.Series({
        "mean_bps": x.mean()*1e4,
        "std_bps": x.std()*1e4,
        "worst_1pct_bps": ax.quantile(0.99)*1e4,
        "max_abs_bps": ax.max()*1e4
    })

stab = pd.concat([
    stability_metrics(df["usdt_b"]).rename("USDT_Binance"),
    stability_metrics(df["usdc_b"]).rename("USDC_Binance"),
    stability_metrics(df["usdt_k"]).rename("USDT_Kraken"),
    stability_metrics(df["usdc_k"]).rename("USDC_Kraken")
], axis=1)

stab
```

```python
ranking = stab.loc["worst_1pct_bps"].sort_values()
ranking
```

```python
# 5.2
THRESH_BPS = 5
THRESH = THRESH_BPS / 1e4
def recovery_time(series, threshold):
    mask = series.abs() > threshold
    groups = (mask != mask.shift()).cumsum()

    durations = mask.groupby(groups).sum()
    return durations[durations > 0]

recovery = {}

for col in df.columns:
    exc = recovery_time(df[col], THRESH)

    recovery[col] = {
        "avg_min": exc.mean(),
        "median_min": exc.median(),
        "p95_min": exc.quantile(0.95)
    }

recovery_df = pd.DataFrame(recovery).T
recovery_df
```
